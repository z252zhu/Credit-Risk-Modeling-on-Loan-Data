Implement a function **WoE_discrete** to automate calculations for discrete variables

```python
def WoE_discrete(df, x, y):
    df = pd.concat([df[x], y], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(), 
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[ :, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_nondefault']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_nondefault'] = df['prop_nondefault'] * df['n_obs']
    df['n_default'] = (1 - df['prop_nondefault']) * df['n_obs']
    df['prop_n_nondefault'] = df['n_nondefault'] / df['n_nondefault'].sum()
    df['prop_n_default'] = df['n_default'] / df['n_default'].sum()
    df['WoE'] = np.log(df['prop_n_nondefault'] / df['prop_n_default'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_nondefault'] = df['prop_n_nondefault'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_nondefault'] - df['prop_n_default']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

Implement a function **plot_by_WoE** to visualize the results

```python
def plot_by_WoE(df_WoE, rotation_x_axis_label = 0, if_xticks = True):
    x = np.array(df_WoE.iloc[: , 0].apply(str))
    y = df_WoE['WoE']
    plt.figure(figsize = (18, 6))
    plt.plot(x, y, marker = 'o', linestyle = '--', color = 'k')
    plt.xticks(visible = if_xticks)
    plt.xlabel('Variable: {}'.format(df_WoE.columns[0]))
    plt.ylabel('Weight of Evidence')
    plt.title('Weight of Evidence by Variable: {}'.format(df_WoE.columns[0]))
    plt.xticks(rotation = rotation_x_axis_label)
    plt.grid()
```

### Preprocessing continuous variables

Implement a function **WoE_ordered_continuous** to automate calculations for continuous variables

```python
def WoE_ordered_continuous(df, x, y):
    df = pd.concat([df[x], y], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False, observed = False)[df.columns.values[1]].agg('count'), 
                    df.groupby(df.columns.values[0], as_index = False, observed = False)[df.columns.values[1]].agg('mean')], axis = 1)
    df = df.iloc[ :, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_nondefault']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_nondefault'] = df['prop_nondefault'] * df['n_obs']
    df['n_default'] = (1 - df['prop_nondefault']) * df['n_obs']
    df['prop_n_nondefault'] = df['n_nondefault'] / df['n_nondefault'].sum()
    df['prop_n_default'] = df['n_default'] / df['n_default'].sum()
    df['WoE'] = np.log(df['prop_n_nondefault'] / df['prop_n_default'])
    df['diff_prop_nondefault'] = df['prop_n_nondefault'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_nondefault'] - df['prop_n_default']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
