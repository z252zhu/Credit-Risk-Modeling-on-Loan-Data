## Probability Default Model Estimation

### Import Libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score
import scipy.stats as stat
```

### Logistic Regression with P-values of the Coefficients

Calculate the p-value of each model coefficient since the calculation is not included in original LogisticRegression function from Scikit-Learn library.

```python
class LogisticRegression_p_values:
    def __init__(self, *args, **kwargs):
        self.model = LogisticRegression(*args, **kwargs)
    
    def fit(self, X, y):
        self.model.fit(X, y)
        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_

        # calculate p_values for each coefficient
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom, (X.shape[1], 1)).T
        F_ij = np.dot((X / denom).T, X)
        F_ij_float64 = F_ij.astype(np.float64)
        Cramer_Rao = np.linalg.inv(F_ij_float64)
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.coef_[0] / sigma_estimates
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores]
        self.p_values = np.append(np.nan, np.round(p_values, 3))
```

### Determing the Cut-offs

Retrieve the variable **false_pos_rate**, **true_positive_rate**, and **threshold** in the model validation section

```python
false_positive_rate.shape
```

```python
true_positive_rate.shape
```

```python
threshold.shape
```

Concate the three variables into a dataframe, **df_cutoffs**

```python
df_cutoffs = pd.concat([pd.DataFrame(threshold), pd.DataFrame(false_positive_rate), pd.DataFrame(true_positive_rate)], axis = 1)
df_cutoffs.columns = ['threshold', 'false_positive_rate', 'true_positive_rate']
df_cutoffs.loc[0, 'threshold'] = 1 - 1 / np.power(10, 10)
df_cutoffs.head(10)
```

Calculate the score based on the probability of non-default using
$$
score\_from\_prob = \left( \ln{ \left( \frac{prob\_non\_default}{1 - prob\_non\_default} \right)} - min\_coef\_sum \right) * \frac{max\_score - min\_score}{max\_coef\_sum - min\_coef\_sum} + min\_score
$$

```python
df_cutoffs['Score'] = (np.log(df_cutoffs['threshold'] / (1 - df_cutoffs['threshold'])) - min_coefficients_sum) * (max_score - min_score) / (max_coefficients_sum - min_coefficients_sum) + min_score
df_cutoffs.loc[0, 'Score'] = 850
df_cutoffs.head(10)
```

```python
df_cutoffs.tail(10)
```

Calculate the corresponding number of approved cases based on the threshold variable, **threshold** 

```python
def number_of_approved(prob):
    return np.where(df_actual_predicted_prob['y_hat_test_prob'] >= prob, 1, 0).sum()
```

```python
df_cutoffs['# Approved'] = df_cutoffs['threshold'].apply(number_of_approved)
df_cutoffs['# Rejected'] = df_actual_predicted_prob['y_hat_test_prob'].shape[0] - df_cutoffs['# Approved']
df_cutoffs['Approval Rate'] = df_cutoffs['# Approved'] / df_actual_predicted_prob['y_hat_test_prob'].shape[0]
df_cutoffs['Rejection Rate'] = 1 - df_cutoffs['Approval Rate']
df_cutoffs.head(10)
```
